```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

import mglearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import shutup

shutup.please()
```

```python
df_gmh = pd.read_csv('gmh.csv')

gmh_cols = [
    "record_id","gender",
    "age",
    "daily_gaming_hours",
    "game_genre",
    "primary_game",
    "gaming_platform",
    "sleep_hours",
    "sleep_quality",
    "sleep_disruption_frequency",
    "continued_despite_problems",
    "eye_strain",
    "back_neck_pain",
    "weight_change_kg",
    "exercise_hours_weekly",
    "social_isolation_score",
    "face_to_face_social_hours_weekly",
    "monthly_game_spending_usd",
    "years_gaming",
    "gaming_addiction_risk_level"
]
game_names = [
    "Dota 2",
    "Elden Ring",
    "StarCraft II",
    "World of Warcraft",
    "Civilization VI",
    "Mobile Legends",
    "League of Legends",
    "Final Fantasy XIV",
    "Cyberpunk 2077",
    "Elder Scrolls Online",
    "Skyrim",
    "Fortnite",
    "Clash of Clans",
    "Warzone",
    "Genshin Impact",
    "CS:GO",
    "PUBG Mobile",
    "Call of Duty",
    "Age of Empires",
    "PUBG",
    "Valorant",
    "Overwatch",
    "Candy Crush",
    "Apex Legends"
]
gender = list(df_gmh['gender'].unique())
platform_names = list(df_gmh['gaming_platform'].unique())
sleep_qual = list(df_gmh['sleep_quality'].unique())
sleep_distf= list(df_gmh['sleep_disruption_frequency'].unique())
awf = list(df_gmh['academic_work_performance'].unique())
mood_state= list(df_gmh['mood_state'].unique())
mood_swing_frequency= list(df_gmh['mood_swing_frequency'].unique())
withdrawal_symptoms= [True, False]
lf_o_i = [True, False]
continued_despite_problems = [True, False]
eye_strain = [True, False]
back_neck_pain = [True, False]
gaming_addiction_risk_level= list(df_gmh['gaming_addiction_risk_level'].unique())

df_gmh["grades_gpa"].fillna(round(df_gmh["grades_gpa"].mean(),2), inplace=True)
df_gmh["work_productivity_score"].fillna(round(df_gmh["work_productivity_score"].mean(),2), inplace=True)

#print(df_gmh.isnull().sum())

print(gender)
```

```python
df_gmh['gender']=df_gmh['gender'].replace(to_replace=["Male","Female","Other"], value=[1,2,3]).astype(int)
df_gmh['game_genre']=df_gmh['game_genre'].replace(to_replace=["MOBA","RPG","MMO","Battle Royale","Strategy","Mobile Games","FPS"], value=[1,2,3,4,5,6,7])
df_gmh['primary_game'] = df_gmh['primary_game'].replace(to_replace=game_names, value=list(np.arange(1,len(game_names)+1)))
df_gmh['gaming_platform'] = df_gmh['gaming_platform'].replace(to_replace=platform_names, value=list(np.arange(1,len(platform_names)+1)))
df_gmh['sleep_quality'] = df_gmh['sleep_quality'].replace(to_replace=sleep_qual, value=list(np.arange(1,len(sleep_qual)+1)))
df_gmh['sleep_disruption_frequency'] = df_gmh['sleep_disruption_frequency'].replace(to_replace=sleep_distf, value=list(np.arange(1,len(sleep_distf)+1)))
df_gmh['academic_work_performance'] = df_gmh['academic_work_performance'].replace(to_replace=awf, value=list(np.arange(1,len(awf)+1)))
df_gmh['mood_state'] = df_gmh['mood_state'].replace(to_replace=mood_state, value=list(np.arange(1,len(mood_state)+1)))
df_gmh['mood_swing_frequency'] = df_gmh['mood_swing_frequency'].replace(to_replace=mood_swing_frequency, value=list(np.arange(1,len(mood_swing_frequency)+1)))
df_gmh['withdrawal_symptoms'] = df_gmh['withdrawal_symptoms'].replace(to_replace=withdrawal_symptoms, value=[1,0])
df_gmh['loss_of_other_interests'] = df_gmh['loss_of_other_interests'].replace(to_replace=lf_o_i, value=[1,0])
df_gmh['continued_despite_problems'] = df_gmh['continued_despite_problems'].replace(to_replace=continued_despite_problems, value=[1,0])
df_gmh['eye_strain'] = df_gmh['eye_strain'].replace(to_replace=eye_strain, value=[1,0])
df_gmh['back_neck_pain'] = df_gmh['back_neck_pain'].replace(to_replace=back_neck_pain, value=[1,0])
df_gmh['gaming_addiction_risk_level'] = df_gmh['gaming_addiction_risk_level'].replace(to_replace=gaming_addiction_risk_level, value=list(np.arange(1,len(gaming_addiction_risk_level)+1)))

df_gmh
```

```python
X = df_gmh[gmh_cols].drop(columns=["record_id","gender"])
y = df_gmh['gender']

knn= KNeighborsClassifier(n_neighbors=22)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state = 8)
knn.fit(X_train, y_train)

print("train accuracy = ", knn.score(X_train, y_train))
print("test accuracy = ", knn.score(X_test,y_test))
```
